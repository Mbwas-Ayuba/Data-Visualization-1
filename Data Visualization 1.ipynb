# Data Visualization

- Visual representation of data to discover trends and patterns.

- Helps in storytelling and decision-making.

## Types of Data Visualization

- Categorical: Bar chart, Pie chart, etc.

- Numerical: Histogram, Box plot, etc.

- Relationship: Scatter plot, Line plot, Heatmap, Pair plot, etc.

## Data Visualization Libraries

- Matplotlib
- Seaborn


### Matplotlib

```python
import pandas as pd

import matplotlib.pyplot as plt
```

- Figure: The entire canvas that holds all subplots
- Axes: A single plot

```python
# Create a figure with a single axes

x = [1, 2, 3]

y = [4, 5, 6]

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("Basic Plot")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")

plt.show() #to display the plot
```

```python
#create a figure with multiple subplots

fig, axes = plt.subplots(1, 2)
axes[0].plot(x, y)
axes[0].set_title("Left Plot")
axes[1].plot(x, y)
axes[1].set_title("Right Plot")
plt.show()
```

### Exploratory Data Analysis (EDA)

- Univariate Analysis: Examining one variable at a time

```python
df["release_year"].value_counts()
```

What are the number of netflix titles released per year? Is there any trend?

```python
titles_per_year = df.groupby("release_year")["title"].count().reset_index()

titles_per_year
```

```python
plt.plot(titles_per_year["release_year"], titles_per_year["title"], color='navy', marker='o')
plt.title("Number of Netflix Titles Released per Year")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")
plt.show()
```

```python
titles_per_country = df.groupby("country")["title"].count().reset_index()

titles_per_country
```

```python
type_counts = df["type"].value_counts()
```

```python
type_counts
```

```python
type_counts.index
```

```python
plt.pie(type_counts, labels=type_counts.index, autopct='%1.2f%%', colors=["navy", "lightblue"])
plt.title("Distribution of Movies and TV Shows")
plt.show()
```

```python
rating_count = df["rating"].value_counts()
```

```python
df["rating"] = df["rating"].replace({"74 min":"Unknown", "84 min": "Unknown", "66 min":"Unknown"})
```

```python
rating_count.values
```

```python
rating_count.index
```

```python
plt.bar(rating_count.index, rating_count.values)
plt.xticks(rotation=90)
plt.title("Number of titles per Ratings")
plt.xlabel("Ratings")
plt.ylabel("Number of titles")
plt.show()
```

```python
plt.hist(df["release_year"], bins=30)
plt.title("Distribution of release years")
plt.show()
```
