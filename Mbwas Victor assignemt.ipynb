```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set
```

```python
df = pd.read_csv("order_details.csv")

df
```

```python
df1 = pd.read_csv("orders.csv")

df1
```

```python
df3 = pd.merge(df, df1, on='order_id', how='inner')

df3
```

```python
df2 = pd.read_csv("pizzas.csv")

df2
```

```python
df4 = pd.read_csv("pizza_types.csv", encoding="latin1")

df4
```

```python
df5 = pd.merge(df2, df4, on='pizza_type_id')

df5
```

```python
data = pd.merge(df3, df5, on='pizza_id', how='inner')


data
```

# 1.	What is the total revenue/sales?

```python
data.info()
```

```python
data['revenue'] = data['quantity']*data['price']

data['revenue']
```

```python
data.head()
```

```python
total_revenue = data["revenue"].sum()

total_revenue
```

# 2.	Find the total quantity sold.

```python
total_quantity_sold = data["quantity"].sum()

total_quantity_sold
```

The total quantity of pizzas sold by category

```python
quantity_of_pizzas_by_category = data.groupby("category")["quantity"].sum()

quantity_of_pizzas_by_category
```

```python
plt.bar(quantity_of_pizzas_by_category.index, quantity_of_pizzas_by_category.values, color="navy", width=0.5)
plt.xticks(rotation=90)
plt.title("Total Quantity of Category Sold")
plt.xlabel("Category")
plt.ylabel("Quantity")
plt.show()
```

- Recommendation:
It can be seen from the figure above that classic pizza has the highest quantity sold, while chicken sold the least. I recommend that more effort should be channelled towards improving chicken for it to sale more

# 3.	Find the total orders

```python
total_orders = data["order_details_id"].nunique()

total_orders
```

Total order of pizza type

```python
top6_pizzas_order = data.groupby("pizza_type_id")["order_details_id"].count().nlargest(6)

top6_pizzas_order
```

```python
plt.barh(top6_pizzas_order.index, top6_pizzas_order.values, height=0.6, color="tomato")
plt.title("Top 6 Pizzas")
plt.ylabel("Pizza")
plt.xlabel("Total Order")
plt.xticks(rotation=45)
plt.show()
```

# 6.	What are the peak hours of sales?

```python
data.info()
```

```python
data["time"] = pd.to_datetime(data["time"])

data["time"]
```

```python
data["hour"] = data["time"].dt.hour
```

```python
data.head()
```

```python
data["sales"] = data["quantity"]*data["price"]
```

```python
data.head()
```

```python
hourly_sales = data.groupby("hour")["sales"].sum().reset_index()

hourly_sales
```

```python
peak_hour_of_sales =hourly_sales.loc[hourly_sales["sales"].idxmax()]

peak_hour_of_sales
```

# 7.	Find the total sales made on each day of the week. Which day of the week is when sales are made the most?

```python
data.info()
```

```python
data["date"] = pd.to_datetime(data["date"])
```

```python
data.info()
```

```python
data["day"] = data["date"].dt.day_name()
```

```python
data.head()
```

```python
sales_per_day =data.groupby("day")["sales"].sum().reset_index()

sales_per_day
```

```python
most_sales_day = sales_per_day.loc[sales_per_day["sales"].idxmax()]

most_sales_day
```

```python
sales_per_day.index
```

```python
sales_per_day.values
```

```python
plt.plot(sales_per_day.index, sales_per_day.values, marker="o")
plt.title("Total Sales Per Day")
plt.xlabel("Days of the Week")
plt.ylabel("Total Sales")
plt.show()
```

* Friday has the highest sales while Sunday has the lowest sales. I recommend more effort to be put inplace to improve sales for Sunday as well as others to be like Friday

# 8.	Find the top 5 bestselling pizzas

```python
 data.head()
```

```python
top5_pizzas = data.groupby("pizza_type_id")["quantity"].sum().nlargest(5)

top5_pizzas
```
